```python
import pandas as pd
df_captions = pd.read_csv("/content/drive/Shareddrives/IA avanzada/Reto_ArcaContinental/Products/image_data_depurado.csv")
df_captions.columns
```

```python
# Obtener los nombres de las características en el conjunto de datos
feature_names = dataset["train"].features.keys()

# Imprimir los nombres de las características
print(feature_names)
```

```python
def map_captions(example):
    # Obtener la ruta del archivo de la imagen actual
    image_path = example["image"].filename

    # Limpia la ruta del archivo para obtener el nombre de archivo
    cleaned_image_name = image_path.replace(parte_a_eliminar, "")

    # Buscar el caption correspondiente en el DataFrame
    matching_captions = df_captions[df_captions["ImageName"] == cleaned_image_name]["Nueva"].values

    if len(matching_captions) > 0:
        # Si se encontraron captions, tomar el primero
        caption = matching_captions[0]
    else:
        # Si no se encontraron captions, asignar un valor por defecto o manejar la situación según tus necesidades
        caption = "Sin caption disponible"

    example["caption"] = caption
    return example
```

```python
dataset["train"] = dataset["train"].map(map_captions)
```

```python
dataset = dataset.with_format("torch")
dataset.set_format(type='torch', columns=['image', 'caption'])
```
