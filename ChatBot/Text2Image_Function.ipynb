```python
from diffusers import DiffusionPipeline
import torch

def Crear_Imagen(prompt):
  pipeline = DiffusionPipeline.from_pretrained("GabrielGaleana34/sd-model-finetuned", torch_dtype=torch.float16).to("cuda")
  negative_prompt = "poor details, deformed, disfigured"
  generator = torch.Generator(device="cuda")
  image = pipeline(prompt=prompt, negative_prompt=negative_prompt, height=1024, width=1024, num_inference_steps=20, guidance_scale=7.5, cross_attention_kwargs={"scale": 0.5}, generator=generator).images[0]
  image.save('imagen_creada.png')



```
